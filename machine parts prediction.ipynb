```python
#decision tree
dt = DecisionTreeRegressor()
dt.fit(xtrain, ytrain)
dt.predict(xtest)
dt.score(xtest,ytest)
```

```python
#Gradient boosting
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor()
gb.fit(xtrain, ytrain)
gb.predict(xtest)
gb.score(xtest,ytest)
```

```python
#adaboost and xgboost
from sklearn.ensemble import AdaBoostRegressor
from xgboost import XGBRegressor
```

```python
ada=AdaBoostRegressor()
ada.fit(xtrain,ytrain)
ada.predict(xtest)
ada.score(xtest,ytest)
```

```python
# prompt: use the ensembel
# \

from sklearn.ensemble import VotingRegressor
# Create a voting regressor
ensemble = VotingRegressor(estimators=[('xgb',xgb),('ada',ada)])

# Fit the voting regressor to the training data
ensemble.fit(xtrain, ytrain)

# Predict the target variable for the test data
y_pred = ensemble.predict(xtest)

# Evaluate the performance of the voting regressor
print('Voting Regressor Score:', ensemble.score(xtest, ytest))
```

#Working And Application

Naive Bayes:
Naive Bayes is a family of probabilistic algorithms that make strong independence assumptions between features. Despite its simplicity, it has been quite successful in various applications. The "naive" assumption is that the presence of a particular feature in a class is independent of the presence of other features.

Working of Naive Bayes:
Assumption of Independence:

Naive Bayes assumes that the features used to describe an observation are independent, which simplifies the computation of probabilities.
Probability Calculation:

It calculates the probability of a given instance belonging to a particular class based on the probabilities of each feature occurring in that class.
Decision Rule:

The class with the highest probability is considered the prediction.
Applications of Naive Bayes:
Text Classification:

Spam filtering is a common application where Naive Bayes is used to classify emails as spam or non-spam.
Sentiment Analysis:

Naive Bayes can be used for sentiment analysis in natural language processing to determine the sentiment of a piece of text, whether it's positive, negative, or neutral.
Medical Diagnosis:

In healthcare, Naive Bayes can be applied to diagnose certain medical conditions based on patient symptoms.
Document Categorization:

It is used for categorizing documents into predefined categories, such as news articles into topics.
Recommendation Systems:

Naive Bayes can be used in recommendation systems to recommend items based on user preferences.
Credit Scoring:

It is applied in credit scoring to assess the creditworthiness of individuals.
Fraud Detection:

Naive Bayes can be used to detect fraudulent activities, such as credit card fraud, by analyzing patterns of normal and abnormal behavior.
Weather Prediction:

In meteorology, Naive Bayes can be applied to predict weather conditions based on various features.
Biomedical Applications:

In bioinformatics, Naive Bayes is used for gene expression analysis and protein classification.
